```python
# ============================================
# MOVIE RAG CHATBOT (GOOGLE COLAB)
# ============================================

# Install dependencies
!pip -q install sentence-transformers faiss-cpu pandas

# ============================================
# IMPORTS
# ============================================

import numpy as np
import pandas as pd
import faiss

from sentence_transformers import SentenceTransformer

# ============================================
# MOVIE KNOWLEDGE BASE
# ============================================

movies = pd.read_csv("/content/movies.csv")

# ============================================
# CREATE DATAFRAME
# ============================================

df = pd.DataFrame(movies)

print("Total Movies:", len(df))
display(df.head())

# ============================================
# LOAD EMBEDDING MODEL
# ============================================

print("Loading embedding model...")

model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

# ============================================
# CREATE EMBEDDINGS
# ============================================

documents = (
    df["title"] +
    ". Genre: " +
    df["genre"] +
    ". Description: " +
    df["description"]
).tolist()

embeddings = model.encode(
    documents,
    convert_to_numpy=True
)

embeddings = embeddings.astype("float32")

print("Embedding Shape:", embeddings.shape)

# ============================================
# CREATE FAISS INDEX
# ============================================

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(embeddings)

print("FAISS Index Created")
print("Total Vectors:", index.ntotal)

# ============================================
# RAG RETRIEVAL FUNCTION
# ============================================

def retrieve_movies(query, top_k=3):

    query_embedding = model.encode(
        [query],
        convert_to_numpy=True
    ).astype("float32")

    distances, indices = index.search(
        query_embedding,
        top_k
    )

    results = []

    for idx in indices[0]:
        results.append(df.iloc[idx])

    return results

# ============================================
# CHATBOT FUNCTION
# ============================================

def movie_chatbot(query):

    results = retrieve_movies(query, top_k=3)

    response = "\n🎬 Recommended Movies:\n\n"

    for i, movie in enumerate(results, start=1):

        response += (
            f"{i}. {movie['title']}\n"
            f"   Genre: {movie['genre']}\n"
            f"   Description: {movie['description']}\n\n"
        )

    return response

# ============================================
# TEST EXAMPLES
# ============================================

print(movie_chatbot("space exploration movies"))

print(movie_chatbot("dreams and mind"))

print(movie_chatbot("action and revenge"))

# ============================================
# INTERACTIVE CHAT LOOP
# ============================================

print("\n==============================")
print("MOVIE RAG CHATBOT")
print("Type 'exit' to stop")
print("==============================")

while True:

    query = input("\nAsk me about movies: ")

    if query.lower() == "exit":
        print("Goodbye!")
        break

    answer = movie_chatbot(query)

    print(answer)
```
