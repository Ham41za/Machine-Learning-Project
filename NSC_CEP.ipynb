```python
# ⚙️ Step 0: Install Dependencies
!pip install mlxtend

# 📚 Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.cluster import KMeans
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# 📁 Step 2: Load Dataset
df = pd.read_csv('/content/Botnet dataset.csv')  # Your provided path
print(f"Dataset shape: {df.shape}")
df.head()
```

```python
# 🧹 Step 3: Data Preprocessing (with NaN handling)
X = df.drop(columns=['label'])  # Replace 'label' if it's named differently
y = df['label']

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Normalize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

```python
# 🎯 Step 4: Feature Selection (Top 25 features using Mutual Information)
selector = SelectKBest(score_func=mutual_info_classif, k=25)
X_selected = selector.fit_transform(X_scaled, y)
selected_features = X.columns[selector.get_support()]
print("✅ Selected Top 25 Features:")
print(selected_features.tolist())
```

```python
# 🌲 Step 5: Random Forest Classification
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print("📊 Classification Report:")
print(classification_report(y_test, y_pred))
print("🧮 Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# 🔍 Step 6: Association Rule Mining on Binary Presence of Top Features
binary_df = df[selected_features].fillna(0).astype(bool)

transactions = binary_df.values.tolist()
te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)
df_te = pd.DataFrame(te_data, columns=te.columns_)

frequent_itemsets = apriori(df_te, min_support=0.1, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

print("📌 Top 10 Association Rules (Sorted by Lift):")
rules.sort_values("lift", ascending=False).head(10)
```

```python
# 🧠 Step 7: K-Means Clustering (2 Clusters: Botnet vs Benign)
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X_selected)

df['Cluster'] = clusters
print("📈 Cluster vs True Label Distribution:")
print(df[['label', 'Cluster']].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.decomposition import PCA
from wordcloud import WordCloud
import numpy as np
import pandas as pd

# Ensure selected_features is array
selected_features_array = np.array(selected_features)

# ✅ 1. Feature Importances - Enhanced
importances = rf.feature_importances_
indices = np.argsort(importances)
plt.figure(figsize=(10, 8))
sns.barplot(x=importances[indices], y=selected_features_array[indices], palette="coolwarm")
plt.title("🔍 Feature Importances from Random Forest")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()

# ✅ 2. Confusion Matrix - Heatmap
plt.figure(figsize=(6, 5))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, display_labels=["Benign", "Botnet"], cmap="YlGnBu", values_format='d')
plt.title("📊 Confusion Matrix - Random Forest")
plt.grid(False)
plt.tight_layout()
plt.show()

# ✅ 3. Association Rule Mining - Lift vs Confidence Bubble Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=rules, x="confidence", y="lift", size="support", hue="support", sizes=(30, 300), palette="Spectral", legend="brief")
plt.title("📌 Association Rules: Lift vs Confidence")
plt.xlabel("Confidence")
plt.ylabel("Lift")
plt.legend(title="Support", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# ✅ 4. K-Means Clustering - PCA 2D Scatter (Color by Cluster)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_selected)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['Cluster'], palette="Set2", style=y, s=60)
plt.title("🧠 K-Means Clustering (PCA Projection)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster (style: true label)", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# ✅ 5. Word Cloud of Selected Features (Permissions)
permissions_text = " ".join(selected_features.tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='tab10').generate(permissions_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("☁️ Word Cloud of Selected Permissions (Top 25 Features)")
plt.tight_layout()
plt.show()

# ✅ 6. Cluster vs True Label Comparison
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Cluster", hue="label", palette="Set3")
plt.title("📈 Cluster Distribution vs True Labels")
plt.xlabel("Cluster Assigned by K-Means")
plt.ylabel("Sample Count")
plt.legend(title="True Label (0=Benign, 1=Botnet)")
plt.tight_layout()
plt.show()
```
